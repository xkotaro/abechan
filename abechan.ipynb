```python
import MeCab

def tokenize(text):
    wakati = MeCab.Tagger('-O wakati -d /usr/local/lib/mecab/dic/mecab-ipadic-neologd')
    wakati.parse("")
    return wakati.parse(text).strip().split()
```

## 差を計算

文書ベクトル同士のコサイン類似度を計算

```python
D_hiroshima_nagasaki_2020 = np.dot(vec_hiroshima_2020,vec_nagasaki_2020) /\
(np.linalg.norm(vec_hiroshima_2020) * np.linalg.norm(vec_nagasaki_2020))
                                                                           
D_hiroshima_nagasaki_2019 =np.dot(vec_hiroshima_2019,vec_nagasaki_2019) /\
(np.linalg.norm(vec_hiroshima_2019) * np.linalg.norm(vec_nagasaki_2019))

D_hiroshima_2020_2019 =np.dot(vec_hiroshima_2020,vec_hiroshima_2019) /\
(np.linalg.norm(vec_hiroshima_2020) * np.linalg.norm(vec_hiroshima_2019))

D_nagasaki_2020_2019 =np.dot(vec_nagasaki_2020,vec_nagasaki_2019) /\
(np.linalg.norm(vec_nagasaki_2020) * np.linalg.norm(vec_nagasaki_2019))
                                      
```

```python
result = {'2019年の広島と長崎の挨拶文の差': D_hiroshima_nagasaki_2020,
          '2020年の広島と長崎の挨拶文の差': D_hiroshima_nagasaki_2019,
          '2019年と2020年の広島の挨拶文の差': D_hiroshima_2020_2019,
          '2019年と2020年の長崎の挨拶文の差': D_nagasaki_2020_2019}
```
